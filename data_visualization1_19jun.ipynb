```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([12,32,56,76,78,89])
y=np.array([13,35,57,78,79,81])
z=np.array([82,76,72,53,34,14])
```

### line plot

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);  #marker size=ms
plt.grid()

#markers=>,<,^,H,o,*
```

```python
plt.plot(x,z);
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='o',mfc='red',color='black',ls='--',ms=5);  #marker size=ms
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);  #marker size=ms
plt.plot(x,z,marker='o',mfc='red',color='black',ls='--',ms=5); 
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5,label='2023');  #marker size=ms
plt.plot(x,z,marker='o',mfc='yellow',color='black',ls='--',ms=5,label='2024'); 
plt.grid()
plt.legend();
```

```python
#subplot
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);  #marker size=ms
plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='o',mfc='red',color='black',ls='--',ms=5,label='2024');  #marker size=ms
plt.grid()
```
